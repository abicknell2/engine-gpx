```python
from gpkit import Variable, Model, units
```

```python
lam = Variable('\\lambda', 1, 'count/hour')
W = Variable('W', 10, 'hours')
L = Variable('L', 'count')
```

```python
m = Model(L, [L>=lam*W])
```

```python
m.solve()
```

```python
from gpx.primitives import Process
```

```python
p = Process()
```

```python
subs = {}
subs[p.t] = 5*units('hr')
subs[p.stdev] = 1*units('hr')
```

```python
from gpx.manufacturing import QNACell, FabLine
```

```python
cells = [QNACell(p) for i in range(3)]
```

```python
line = FabLine(cells)
```

```python
# subs[line.lam] = 1*units('count/hr')
```

```python
constr = [p, line]
```

```python
import numpy as np
```

```python
cost = line.L + np.sum([c.m for c in cells])
```

```python
m = Model(cost,constr)
```

```python
m.substitutions.update(subs)
```

```python
sol = m.solve()
```
