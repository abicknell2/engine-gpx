## JSON handling for nested objects

```python
import json
```

```python
test_dict = {
    'list' : [1,2,3,4],
    'dict' : {'one' : 1, 'two': 2, 'three':3}
}
```

```python
print(json.dumps(test_dict, indent=4))
```

```python
test_list = [test_dict for i in range(4)]
```

```python
json.dumps(test_list)
```
