```python
class FeederLine(FabLine):
    '''A Short Feeder Line

    The short feeder line comprises a CONWIP line which terminates in a batching process

    Usually there is very weak cost on inventory in feeder lines (unless there is tooling)
    May need to manually add an upper limit
    '''
    def setup(self, cells, return_cells=False, target_cell, **kwargs):
        '''
        ARGS
        ----
        is_batched : boolean
            does the feeder line need a batching process at the end
        batch_qty : int, gpkit.Variable, gpkit.Monomial
            the quantity which has to be batched
        '''
        constraints = super().setup(cells, return_cells)    # get all the regular conwip constraints
        constraints.append(
            target_cell.c2a >= cells[-1].c2d    # model by finding worst case of the input variation
        )
        return constraints
    
    def get_end_buffer(self, sol):
        'get the buffer right before the feed into the mainline'
        W = self.cells[-1].Wq
        lam = self.lam
        return sol['variables'][w]*sol['variables'][lam]
```
