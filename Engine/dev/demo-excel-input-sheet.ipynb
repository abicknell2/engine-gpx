# Demo of Excel Input Sheet

- capture all non-constrained inputs and put to excel with sheetname `#linelab_inputs`

### Imports

```python
import pandas as pd
import json
import numpy as np
```

### Open Json

```python
with open('Inputs_dump.json') as fh:
    inputjson = json.load(fh)
```

```python
# filter the inputs to have only elements with `sign`
filtered = filter(lambda x: 'isConstrained' not in x, inputjson)
filtered2 = list(filter(lambda x: 'Design' not in x['module'], filtered))
```

### Put into pandas

```python
df = pd.DataFrame(filtered2)
```

```python
df
```

```python
# sort the columns
cols = ['name', 'worst', 'likely', 'best', 'unit']
```

```python
df[cols].replace(np.nan, '').set_index('name')
```

```python
df[cols].replace(np.nan, '').set_index('name').to_excel('test_input.xlsx', sheet_name='#linelab_inputs')
```
