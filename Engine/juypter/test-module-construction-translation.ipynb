```python
from gpx.manufacturing import QNACell, FabLine
from gpx.primitives import Process
```

```python
from collections import OrderedDict as OD
```

```python
processes = OD()
cells = OD()

cellnames = list(active_modules['manufacturing'].cells.keys())

gpxvars = active_modules['manufacturing'].gpx_variables
```

```python
for name, proc in active_modules['manufacturing'].processChain.items():
    processes[name] = Process(
        time=gpxvars[name + ' Time'],
        stdev=gpxvars[name + ' Std. Dev.']
    )
    cells[cellnames[proc['cell']-1]] = QNACell(
        process=processes[name]
    )
```

```python
line = FabLine(list(cells.values()))
```

```python
from gpx.non_recurring_costs import ConwipToolingCost, CellCost
```

```python
toolcosts = []
for tool in active_modules['manufacturing'].tools:
    toolcosts.append(ConwipToolingCost(
        list(cells.values()),
        starting_cell=tool['first'],
        ending_cell=tool['last'],
        tool_cost=gpxvars[tool['name'] + ' Tool Cost']
    ))
```

```python
cellcosts = OD()
for name, cell in cells.items():
    cellcosts[name] = CellCost(cell, workstation_cost=gpxvars[name + ' Cell Cost'])
```
