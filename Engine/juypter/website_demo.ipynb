QNA states that the aggregate process coefficient of variation, $\bar{c_s^2}$ is found:
$$ \bar{\tau^2}(\bar{c_s^2} + 1) = \sum x_i\tau_i(c_{si}^2 + 1) $$

Knowing 
$$ \bar{\tau} = \sum x_i\tau_i$$
We can write 
$$ \bar{\tau^2}\bar{c_2^2} = x_1\tau_1^2 c_{s1}^2 + x_2\tau_2^2 c_{s2}^2 + \mathcal{N} $$
where
$$ \mathcal{N} = x_1\tau_1^2 + x_2\tau_2^2 -\left(x_1^2\tau_1^2 + 2x_1x_2\tau_1\tau_2 + x_w^2\tau_2^2 \right) $$

Furthermore, 
$$ x_1,x_2 > 0 \quad \mathrm{and} \quad x_1 + x_2 = 1$$

therefore it can be shown that 
$$ \mathcal{N} > 0 $$
